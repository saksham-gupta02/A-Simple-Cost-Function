# Import Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example-1 : A Simple Cost Function

$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2+x+1
```

```python
# Make Data

x_1 = np.linspace(start=-3,stop=3,num=100)
```

```python
# Plot f(x)

plt.plot(x_1,f(x_1))
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.show()
```

## Slope and Derivatives

```python
def df(x):
    return 2*x +1
```

```python
# Plot the function and derivative

plt.figure(figsize=[15,5])

# Chart 1 : Cost Function
plt.subplot(1,2,1)
plt.title("Simple Cost Function",fontsize=15)
plt.plot(x_1,f(x_1),color='Blue',linewidth = 3)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

# Chart 2 : Derivative

plt.subplot(1,2,2)
plt.title("Derivative of the Cost Function",fontsize=15)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)

plt.grid()
plt.plot(x_1,df(x_1),color='skyblue')
plt.show()
```

```python
# Gradient Descent

new_x = 3
previous_x = 0
step_multiplier = 0.1 # Learning Rate
precision = 0.0001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(50):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier*gradient
    
    x_list.append(new_x)
    slope_list.append(df(new_x))
    
    step_size = abs(new_x - previous_x)
    if step_size < precision:
        print("This Loop ran",n,"times")
        break
    
print("Local Minimum occurs at:",new_x)
print("Slope for df(x) at this point is:",df(new_x))
print("f(x) or cost at this point is:",f(new_x))
```

```python
# Plot the function and derivative

plt.figure(figsize=[20,5])

# Chart 1 : Cost Function
plt.subplot(1,3,1)
plt.title("Simple Cost Function",fontsize=15)
plt.plot(x_1,f(x_1),color='Blue',linewidth = 3)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

values = np.array(x_list)
plt.scatter(x_list,f(values),color='red',alpha=0.4)

# Chart 2 : Derivative

plt.subplot(1,3,2)
plt.title("Derivative of the Cost Function",fontsize=15)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.scatter(x_list,slope_list,color='red',alpha=0.5)
plt.grid()
plt.plot(x_1,df(x_1),color='skyblue')


# Chart 3 : Derivative Close Up

plt.subplot(1,3,3)
plt.title("Derivative of the Cost Function",fontsize=15)
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.scatter(x_list,slope_list,color='red',alpha=0.5,s=300)
plt.grid()
plt.plot(x_1,df(x_1),color='skyblue',linewidth=6,alpha=0.6)


plt.show()
```
