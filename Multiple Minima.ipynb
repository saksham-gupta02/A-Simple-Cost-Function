# Example 2 : Multiple Minima
$$ g(x)=x^4 - 4x^2 + 5 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_2 = np.linspace(-2,2,1000)
```

```python
# Creating g(x) function
def g(x):
    return x**4 - 4*x**2 + 5
```

```python
# Creating derivative of g(x) as dg(x)

def dg(x):
    return 4*x**3 - 8*x
```

```python
# Plotting Function and its slope side by side

plt.figure(figsize=[20,5])

# Chart 1 : Plotting g(x)
plt.subplot(1,2,1)
plt.title('Cost function g(x)',fontsize=10)
plt.plot(x_2,g(x_2),color='blue',linewidth=3)
plt.xlabel('X',fontsize=16)
plt.ylabel('g(x)',fontsize=16)
plt.xlim(-2,2)

# Chart 2 : Plotting dg(x) - Derivative/Slope of g(x)
plt.subplot(1,2,2)
plt.title('Derivative of Cost function g(x)',fontsize=10)
plt.plot(x_2,dg(x_2),color='skyblue',linewidth=2)
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(x)',fontsize=16)
plt.xlim(-2,2)

plt.show()
```

```python
# Creating Gradient Descent Function

def gradient_descent(derivative_func,initial_guess,multiplier=0.02,precision=0.0001):
    
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    
    for n in range(500):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - gradient*multiplier
        
        step_size = abs(new_x-previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        
        if step_size < precision:
            break
    return new_x,x_list,slope_list
```

```python
# Calling Our Gradient Function and getting the values the function returns
local_min,list_x,deriv_list = gradient_descent(dg,0.5,0.02,0.001)
print("Local minimum occurs at:",local_min)
print("Number of steps:",len(list_x))
```

```python
# Plotting the Gradient func result on our Graphs

# Plotting Function and its slope side by side

plt.figure(figsize=[15,5])

# Chart 1 : Plotting g(x)
plt.subplot(1,2,1)
plt.title('Cost function g(x)',fontsize=10)
plt.plot(x_2,g(x_2),color='blue',linewidth=3)
plt.scatter(list_x,g(np.array(list_x)),color='red')
plt.xlabel('X',fontsize=16)
plt.ylabel('g(x)',fontsize=16)
plt.xlim(-2,2)

# Chart 2 : Plotting dg(x) - Derivative/Slope of g(x)
plt.subplot(1,2,2)
plt.title('Derivative of Cost function g(x)',fontsize=10)
plt.plot(x_2,dg(x_2),color='skyblue',linewidth=2)
plt.scatter(list_x,deriv_list,color='red')
plt.grid()
plt.xlabel('X',fontsize=16)
plt.ylabel('dg(x)',fontsize=16)
plt.xlim(-2,2)


plt.show()
```
