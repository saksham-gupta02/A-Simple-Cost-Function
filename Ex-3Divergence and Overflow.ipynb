# Example 3 : Divergence and Overflow

## $$ h(x) = x^5 - 2x^4 + 2 $$

```python
# Import relevant Libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make some Data for the Function
x_3 = np.linspace(-2.5,2.5,1000)

def h(x):
    return x**5 - 2*x**4 + 2

def dh(x):
    return 5*x**4 - 8*x**3
```

```python
def gradient_descent(derivative_func,initial_guess,multiplier = 0.02,precision = 0.001,max_iter = 300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]
    
    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - gradient * multiplier
        
        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))
        
        if(step_size < precision):
            break
        
    return new_x,x_list,slope_list # Returning values as Tuples
    
```

```python
local_min, list_x,deriv_lis = gradient_descent(dh,0.2) # getting values in tuples
```

```python
# Plot Function and derivative side by side

plt.figure(figsize=[15,5])

# Chart 1: Function h(x)

plt.subplot(1,2,1)

plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.plot(x_3,h(x_3),linewidth=3,color='blue',alpha=0.6)
plt.scatter(list_x,h(np.array(list_x)),color = 'red',alpha=0.4)

# Chart 2 : Derivative of the function

plt.subplot(1,2,2)
plt.grid()
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.plot(x_3,dh(x_3),linewidth=3,color='skyblue',alpha=0.6)
plt.scatter(list_x,deriv_lis,color='red',alpha=0.4)

plt.show()
```

## Below is the example of over flow

#### Here our initial guess is -0.2 and as our algo learn it will go down and will run infinitely untill we give some max number of iteration manually

```python
local_min, list_x,deriv_lis = gradient_descent(dh,-0.2,max_iter=71) # getting values in tuples

# Plot Function and derivative side by side

plt.figure(figsize=[15,5])

# Chart 1: Function h(x)

plt.subplot(1,2,1)

plt.xlim(-1.2,2.5)
plt.ylim(-1,4)
plt.plot(x_3,h(x_3),linewidth=3,color='blue',alpha=0.6)
plt.scatter(list_x,h(np.array(list_x)),color = 'red',alpha=0.4)

# Chart 2 : Derivative of the function

plt.subplot(1,2,2)
plt.grid()
plt.xlim(-1,2)
plt.ylim(-4,5)
plt.plot(x_3,dh(x_3),linewidth=3,color='skyblue',alpha=0.6)
plt.scatter(list_x,deriv_lis,color='red',alpha=0.4)

plt.show()
```
